```python
import requests
from bs4 import BeautifulSoup
import re

def get_faculty_data(name, url):
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.content
        soup = BeautifulSoup(html_content, 'html.parser')

        # Extract data from the specified divs
        bio_div = soup.find('div', {'id': 'bio'})

        # Extracted data
        faculty_bio = bio_div.get_text(strip=True) if bio_div else "No bio available"

        # Save data to the same text file
        with open('bios.txt', 'a', encoding='utf-8') as output_file:
            output_file.write(f"\nFaculty Name: {name}\n")
            output_file.write(f"Bio:\n{faculty_bio}\n")

        print(f"Data for {name} has been appended to bios.txt\n")
    else:
        print(f"Failed to retrieve the webpage for {name}. Status code: {response.status_code}")

# Read names and URLs from the text file
file_path = 'bio_urls.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Process each name and URL using regular expressions
for i in range(0, len(lines), 2):
    name_line = lines[i].strip()
    url_line = lines[i + 1].strip()

    if name_line and url_line:
        name_match = re.match(r'^(?P<name>.*?)$', name_line)
        url_match = re.match(r'^URL:(?P<url>https://.*)$', url_line)

        if name_match and url_match:
            name = name_match.group('name')
            url = url_match.group('url')
            get_faculty_data(name, url)
        else:
            print("Invalid line format. Skipping:", name_line, url_line)
    else:
        print("Invalid line format. Skipping:", name_line, url_line)
```
